```python
## independnent and dependent features
X=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
##train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
X_train.head()
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
```

```python
# Skip this - ColumnTransformer will handle test data encoding
```

```python
X_test.head()
```

```python
## Onehot encoding--- ColumnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
ct=ColumnTransformer(
    transformers=[
        ('cat_features', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), [1, 2, 3, 4]),
        ('num_features', 'passthrough', [0, 5])
    ],
    remainder='passthrough'
)
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```

```python
X_test
```

```python
## SVR--Support Vector Regression
from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
## Hyperparameter Tuning using GridSearch CV
```

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)
 
# fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_prediction=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,grid_prediction))
print(mean_absolute_error(y_test,grid_prediction))
```
