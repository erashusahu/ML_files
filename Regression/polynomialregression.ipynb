```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2+x+2+np.random.randn(100,1)
#this is my quadratioc equation with some noise added to it
plt.scatter(x,y,color="g")
plt.xlabel("X-label")
plt.ylabel("y-label")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
sscore=r2_score(y_test,regression_1.predict(x_test))
print(sscore)
```

```python
## lets visualize this model
plt.plot(x_train,regression_1.predict(x_train),color='r')
plt.scatter(x_train,y_train)
plt.xlabel(" X Dataset")
plt.ylabel(" y dataset")
plt.show()
```

```python
#lets apply polynomail transformation 

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
```

```python
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
```

```python
x_train_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(x_train_poly,y_train)
y_pred=regression.predict(x_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
plt.scatter(x_train,regression.predict(x_train_poly))
plt.scatter(x_train,y_train)
plt.show()
```

```python
poly=PolynomialFeatures(degree=3,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)
x_train_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(x_train_poly,y_train)
y_pred=regression.predict(x_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
# prediction of new data
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
X_new_poly
```

```python
y_new=regression.predict(X_new_poly)
plt.plot(X_new,y_new, "r-",linewidth=2,label="New Prediction")
plt.plot(x_train,y_train,"b.",label="Training data")
plt.plot(x_test,y_test, "g",label="Testing Point")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
